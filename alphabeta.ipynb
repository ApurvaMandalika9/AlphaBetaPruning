```python
#######################################
# Alphabeta search
#
# Yoonsuck Choe
# 9/26/2021 - revised 9/28
#
# - This is a really basic implementation.
# - You can consider modifying it to keep track of the solution path
#
#######################################

#######################################
# Constants

MAX = 99999
MIN = -99999

#######################################
# Game tree  : Change the list "Game_tree" to test a different game
'''
                       o MAX
                       |
           ____________|_________
          |                      |
    ______o________           ___o___     MIN
   |      |        |         |       |
   7    __o___    _o_        12    __o__       MAX
       |  |  |   |   |            |     |
       3  8  1   15  9           17     13
'''

#Game_tree = [[7, [3, 8, 1], [15, 9]], [12, [17, 13]]]
Game_tree = [[[5,8],[3,[5,2,7],9,[5,8,1]],[7,[9,10,8]]],[[[3,6,4],6,[10,3]],[[8,13],5,[10,8]]]]

#######################################
# Function definitions

def alphabeta_decision(tree):
  '''
  input: game tree (as a python list)
  return: value of optimal outcome
  '''

  # compact implementation:

  return alphabeta_value(tree,"max", MIN, MAX)


def alphabeta_value(tree, node_type, alpha, beta, indent=""):

  '''
  input: game tree, the type of its root : "max" or "min", alpha, beta
  return: value of optimal outcome
  '''

  if (type(tree) == int):      # Leaf node
    print(indent+"CALL", node_type, "returns", tree)
    return tree

  elif (node_type == "max"):  # "tree"'s root is a MAX node

    max_v = MIN

    print(indent+"CALL max", str(tree), "alpha =", alpha, "beta =", beta)

    for i in tree:
      v = alphabeta_value(i, "min", alpha, beta, indent+"  ")
      max_v = max(max_v, v)
      alpha = max(alpha, v) # updating only alpha as the node type we are dealing with is a MAX node. Beta is left as it is.

      print(indent+"  - max", tree, "returns", max_v, "alpha =", alpha, "beta =", beta)

      if beta <= alpha:
        print(indent+"  - Pruned")
        break

    return max_v

  elif (node_type == "min"):  # "tree"'s root is a MIN node

    min_v = MAX

    print(indent+"CALL min", str(tree), "alpha =", alpha, "beta =", beta)

    for i in tree:
      v = alphabeta_value(i, "max", alpha, beta, indent+"  ")
      min_v = min(min_v, v)
      beta = min(beta, v) # updating only beta as the node type we are dealing with is a MIN node. Alpha is left as it is.

      print(indent+"  - min", tree, "returns", min_v, "alpha =", alpha, "beta =", beta)

      if beta <= alpha: #Base case for pruning
        print(indent+"  - Pruned")
        break

    return min_v

#######################################
# Test

alphabeta_decision(Game_tree)
```
